```python
import pandas as pd
import numpy as np
import io
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from scipy.spatial.distance import cdist
```

```python
df_search_words = pd.read_excel('../data/Search-words-june 2018-2019.xlsx')
```

```python
search_words_list = list(df_search_words['Waarde'])
```

```python
word_emb = {}
```

```python
stop_words_file_nl = open('../data/misc_files/stopwords_dutch.txt', 'r')
stop_words_nl = stop_words_file_nl.read().split('\n')
```

```python
with io.open('../data/word2vec/wikipedia-160.txt', 'r', encoding="utf-8") as infile:
    first_time = 0
    for line in infile:
        if first_time == 0:
            first_time = 1
            continue
        split = line.split()    
        word = split[0]
        embedding = np.array([float(val) for val in split[1:]])
        word_emb[word] = embedding
```

```python
words = []
for word in search_words_list:
    if str(word).replace(' ', '').isalpha():
            words.extend(word.split())
            
words = [word.lower() for word in words]
```

```python
word_emb_vals = list(word_emb.values())
word_emb_keys = list(word_emb.keys())
```

```python
similar_words = []
final_search_words = []
for word in words:
    try:
        src_word_emb = word_emb[word]
    except:
        continue
    dist = cdist([src_word_emb], word_emb_vals)
    indices = np.argsort(dist[0])[1:6]
    sub_lst = [word_emb_keys[ind] for ind in indices]
    similar_words.append(sub_lst)
    final_search_words.append(word)
```

```python
df = pd.DataFrame()
df['search_word'] = final_search_words
df['similar_words'] = similar_words
df.to_csv('similar_words.csv', sep='\t', index=False)
```
